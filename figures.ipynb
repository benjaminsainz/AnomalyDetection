```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
no_normalization = pd.read_csv('benchmark_no_normalization.csv', header=0).drop('dataset', axis=1)
minmax = pd.read_csv('benchmark_minmax_normalization.csv', header=0).drop('dataset', axis=1)
standard = pd.read_csv('benchmark_standard_normalization.csv', header=0).drop('dataset', axis=1)
```

```python
no_normalization_melted = no_normalization.melt(var_name='algorithm', value_name='auc')
minmax_melted = minmax.melt(var_name='algorithm', value_name='auc')
standard_melted = standard.melt(var_name='algorithm', value_name='auc')
```

```python
fig, ax = plt.subplots(figsize=(12, 8), dpi=200)
sns.boxplot(data=no_normalization_melted, x='algorithm', y='auc', palette='pastel')

ax.set_title('No Normalization')
ax.set_xlabel('')
ax.set_ylabel('AUC')
    
plt.tight_layout()
sns.despine(left=True, bottom=True)
fig.savefig('boxplot_no_normalization.png', format='png', dpi=1200)
```

```python
fig, ax = plt.subplots(figsize=(12, 8), dpi=200)
sns.boxplot(data=minmax_melted, x='algorithm', y='auc', palette='pastel')

ax.set_title('MinMax Normalization')
ax.set_xlabel('')
ax.set_ylabel('AUC')
    
plt.tight_layout()
sns.despine(left=True, bottom=True)
fig.savefig('boxplot_minmax.png', format='png', dpi=1200)
```

```python
fig, ax = plt.subplots(figsize=(12, 8), dpi=200)
sns.boxplot(data=standard_melted, x='algorithm', y='auc', palette='pastel')

ax.set_title('Standard Normalization')
ax.set_xlabel('')
ax.set_ylabel('AUC')
    
plt.tight_layout()
sns.despine(left=True, bottom=True)
fig.savefig('boxplot_standard.png', format='png', dpi=1200)
```
